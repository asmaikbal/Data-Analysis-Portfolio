```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Empty DataFrame to store outliers
outliers = pd.DataFrame()

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

 # Find rows where this column has outliers
    col_outliers = df[(numeric_df[col] < lower_bound) | (numeric_df[col] > upper_bound)]

    if not col_outliers.empty:
        print(f"\nOutliers detected in column: {col}")
        print(col_outliers[[col]])

    outliers = pd.concat([outliers, col_outliers])

# Remove duplicates if a row is an outlier in multiple columns
outliers = outliers.drop_duplicates()

print(f"\nTotal rows with at least one outlier: {len(outliers)}")
```

```python
# Cleaning Data, Ensuring relevant columns are numeric
df['DAILY_STRESS'] = pd.to_numeric(df['DAILY_STRESS'], errors='coerce')
df['WORK_LIFE_BALANCE_SCORE'] = pd.to_numeric(df['WORK_LIFE_BALANCE_SCORE'], errors='coerce')

# Selecting only numeric columns for analysis
df_numeric = df.select_dtypes(include=np.number)

# Creating a list for the summary
outlier_satistical_summary = []

# Skim through all numeric columns to calculate outliers
for column in df_numeric.columns:
    data = df_numeric[column].dropna()

    if data.empty:
        continue

    # Calculate Quartiles and IQR (as shown in class)
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    # Calculate the outlier boundaries (Fence/Thresholds)
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    # Count Outliers
    lower_outliers = data[data < lower_fence].count()
    upper_outliers = data[data > upper_fence].count()
    total_outliers = lower_outliers + upper_outliers

    # Add results to summary list created above
    outlier_satistical_summary.append({
        'Column': column,
        'Total Entries': len(data),
        'Lower Fence': f"{lower_fence:.2f}",
        'Upper Fence': f"{upper_fence:.2f}",
        'Total Outliers': total_outliers,
        'Percentage': f"{(total_outliers / len(data) * 100):.2f}%"
    })

df_summary = pd.DataFrame(outlier_satistical_summary)
print(df_summary)
```
